Calculadora del índice de masa corporal

Escribe una función que calcule el IMC ingresado por el usuario/a, es decir, quien lo ejecute deberá ingresar estos datos. Puedes obtener más información de su cálculo en:

-> Índice de masa corporal IMC que es y cómo se calcula.

La función debe clasificar el resultado en sus respectivas categorías

```python
#Ejercicio 1

#Bajo peso: menos de 18,5.
#Peso normal: 18,5-24,9.
#Sobrepeso: 25-29,9.
#Obesidad: más de 30.

def calculadora_imc():
    '''
    Calculadora el Indice de Grasa Corporal
    Para calcular el IMC es necesario dividir el peso en kg de un paciente 
    entre el cuadrado de su altura en metros (peso/(altura**))
    '''
    #Calculadora de Indice de Grasa Corporal.
    #Solicitamos al usuario el peso y la altura
    print("Hola, bienvenido/a a la caaaalculadora de Indice de masa corporal. Cual es tu nombre?")
    print("\n")
    nombre = input("Ingresa tu nombre")
    print(f"Hola!!! {nombre}")
    print("Recuerda de ingresar la altura en metros.")
    print("por ejemplo: forma correcta: 1.75. forma incorrecta: 175") 
    print("\n")
    print("Si no lo has hecho asi, vuelvelo a intentar")
    print("Si deseas salir, escribe 'salir' en cualquier momento.")
    print("\n")
    while True:

        altura = input("Bienvenido/a, indica tu altura en metros.")
        altura = str(altura.replace(',','.'))
        if altura.lower() == "salir":
            print("Has decidido salir de la calculadora.")
            return
        elif float(altura) > 3.00 or float(altura) < 1.00:
            print("Esa altura no es valida")
        else:
            try:
                altura = float(altura)
                print(f"Perfecto!! tu altura es {altura}")
                break
            except ValueError:
                print("Debes ingresar la altura en metros. Forma correcta 1.75. Forma incorrecta = 175")
    while True:     
        peso = input("Ahora indica tu Peso.")
        if peso.lower() == "salir":
            print("Has decidido salir de la calculadora.")
            return
        elif int(peso) > 400:
            print("Es un peso muy elevado, ¿estas seguro? Intentalo de nuevo")
        else:
            try:
                peso = peso.replace(',','.')
                peso = float(peso)
                print(f"Genial!! tu peso es {peso}")
                break
            except ValueError:
                print("Introduzca un numero valido o escribe 'salir'")
        print("\n")
        print(f"Bueno entonces pesas {peso} y mides {altura}")
        print("\n")

    imc = peso/(altura ** 2)
    print(f"Tu IMC es {imc:.2f}")

    if imc < 18.5: #Bajo peso: menos de 18,5.
        print("Segun mis calculos tu indice de masa corporal indica que tienes bajo peso")
    elif 18.5 <= imc < 24.9: #Peso normal: 18,5-24,9
        print("Segun mis calculos tu indice de masa corporal indica que tu peso es normal")
    elif 25 <= imc < 29.9: #Sobrepeso: 25-29,9.
        print("Segun mis calculos tu indice de masa corporal indica que tienes Sobrepeso")
    else: 
        print("Segun mis calculos tu indice de masa corporal indica que tienes obesidad")


        
calculadora_imc()
    

```

Conversor de temperaturas.

Existen diversas unidades de temperatura utilizadas en distintos contextos y regiones. Las más comunes son Celsius (°C), Fahrenheit (°F) y Kelvin (K). También existen otras unidades como Rankine (°Ra) y Réaumur (°Re). 


Selecciona al menos 2 conversores, de modo que al introducir una temperatura devuelva, como mínimo, dos conversiones.


```python
def conversor_temperaturas():

    '''
    Conversor de temperaturas
    Puedes escoger alguno de los tipos de unidades de temperatura
    Introducir el valor, y te dira las conversiones de las demas.
    '''
    print("Bienvenido al conversor de temperaturas!")
    print("Puedes convertir entre Celcius (C), Fahrenheit (F) y Kelvin (K)")
    print("Deberas elegir alguno de ellos")
    print("Si deseas salir, solo escribe 'salir'")
    print("\n")

    #Aqui el usuario debera ingresar la unidad de temperatura
    
    while True:
        elegir_unidad = input("¿Cual es la unidad que deseas convertir? Ingresa 'C' para Celsius, 'F' para Fahrenheit o 'K' para Kelvin").strip().lower()
        if elegir_unidad.lower() == "salir":
            print("Has decidido salir del conversor de temperatura.")
            return
        if elegir_unidad in ['c','celsius','f','fahrenheit','k','kelvin']:
            break    
        else:
            print("No has ingresado correctamente")
            
    
    while True:
        ingresar_numero = input("Ingresa un numero, el valor de la temperatura que deseas convertir: ").replace(',','.')
        if ingresar_numero.lower() == 'salir':
            print("Has decidido salir del conversor de temperatura.")
            return
        try:
            ingresar_numero = float(ingresar_numero)
            print(f"El valor ingresado es: {ingresar_numero}")
            break
        except ValueError:
            print("Ingresaste un numero incorrecto, por favor vuelvelo a ingresar")
            print("\n")

    if elegir_unidad in ['c','celsius']:
        print("Escogiste Celsius")
        fahrenheit = ingresar_numero * (9/5) + 32
        kelvin = ingresar_numero + 273.15
        print(f"Entonces {ingresar_numero} grados Celsius equivale a {fahrenheit:.2f} grados Fahrenheit y a {kelvin:.2f} Kelvin")
    elif elegir_unidad in ['f','fahrenheit']:
        print("Escogiste Fahrenheit")
        celsius = (ingresar_numero - 32) * (5/9)
        kelvin = (ingresar_numero - 32) * (5/9) + 273.15
        print(f"Entonces {ingresar_numero} grados Fahrenheit equivale a {celsius:.2f} grados Celsius y a {kelvin:.2f} Kelvin")
    elif elegir_unidad in ['k','kelvin']:
        print("Escogiste Kelvin")
        celsius = ingresar_numero - 273.15
        fahrenheit = (ingresar_numero - 273.15) * (9/5) + 32
        print(f"Entonces {ingresar_numero} Kelvin equivale a {celsius:.2f} grados Celsius y a {fahrenheit:.2f} grados Fahrenheit")
        
conversor_temperaturas()        
```

Contador de palabras de un texto.

Escribe una función que dado un texto, muestre las veces que aparece cada palabra.

```python
def contador_palabras(texto):
    '''
    En este contador de palabras vamos a analizar
    cuantas palabras se repiten en 
    las siguientes canciones o textos.
    '''

    texto = texto.lower()

    palabras = ""
    for letra in texto:
        if letra.isalpha() or letra.isspace(): 
            palabras += letra
            
            
    
    #Dividir el texto en palabras
    lista_palabras = palabras.split() #funcion split para dividir
    
    #Se crea diccionario donde cada palabra(clave) se realizara el contaje(valor)
    conteo = {} 
    for palabra in lista_palabras:
        if palabra in conteo:
            conteo[palabra] += 1
        else:
            conteo[palabra] = 1
    
    #Mostrar el conteo de palabras ordenadas alfabeticamente
    print("Conteo de palabras:")
    for palabra in sorted(conteo): 
        print(f"{palabra}: {conteo[palabra]}") 

# Canciones

crimen_cerati = """La espera me agotó 
No sé nada de vos
Dejaste tanto en mí
En llamas me acosté
Y en un lento degradé
Supe que te perdí"""

demoliendo_hoteles_charly_garcia = """Yo, que crecí con Videla
Yo, que nací sin poder
Yo, que luché por la libertad
Pero nunca la pude tener

Yo, que viví entre fachistas
Yo, que morí en el altar
Yo, que nací con los que estaban bien
Pero, a la noche, estaba todo mal

Hoy paso el tiempo
Demoliendo hoteles
Mientras los plomos juntan los cables
Cazan rehenes"""



contador_palabras(crimen_cerati)
```

Diccionario inverso.


Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesita a la inversa, es decir, intercambiados las claves y valores. 


Los valores y claves en el diccionario original son únicos; si éste no es el caso, la función debería imprimir un mensaje de advertencia.

```python
def invertir_diccionario(diccionario):
    """
    Inverte un diccionario, intercambiando claves y valores.
    Si hay claves o valores repetidos, imprime un mensaje de advertencia.
    """
    dic_invertido = {}
    dic_claves = set()  
    dic_valores = set()  

    for clave, valor in diccionario.items():
        if clave in dic_claves or valor in dic_valores:
            print(f"Advertencia: clave '{clave}' o valor '{valor}' ya existe en el diccionario original.")
        else:
            dic_claves.add(clave)
            dic_valores.add(valor)
            dic_invertido[valor] = clave

    return dic_invertido

diccionarios = ({'a': 1, 'b': 2, 'c': 3})
diccionario2 = ({'x': "apple", 'y': "banana", 'z': "banana"})

invertir_diccionario(diccionario2)

```
